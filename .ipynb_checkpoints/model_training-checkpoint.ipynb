```python
#Library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing dataset from kagle

# link: https://www.kaggle.com/code/sagarlokesh/diabetes-prediction-logistic-regression

data = pd.read_csv('diabetes.csv') # importing
data.head(10) # visualization our dataset
```

```python
# verifying the variables from dataset, correlation between it

data.corr().round(4)   # corr() = correlation
```

```python
data.columns # colunms
```

```python
# getting only necessaries columns for analysis (after correlation analysis)

data_2 = data[['Pregnancies', 'Glucose', 'BMI', 'Age', 'Outcome']]
```

```python
# Correlation between choose variables
sns.heatmap(data_2.corr(),
            annot=True,
            vmax=1,
            vmin=-1,
            cmap='BrBG',
            linewidths=1,
            xticklabels=['Gestaes', 'Glicose', 'IMC', 'Hist. Familiar', 'Idade', 'Resultado'],
            yticklabels=['Gestaes', 'Glicose', 'IMC', 'Hist. Familiar', 'Idade', 'Resultado'])
plt.savefig('imgs/Correlation')
```

```python
data_2.shape # size of dataframe
```

```python
# Machine learning libraries

from sklearn.metrics import accuracy_score, confusion_matrix     # metrics for our algoritm
from sklearn.model_selection import train_test_split             # we need slip dataset in train and test
from sklearn.neighbors import KNeighborsClassifier               # classifier used 
```

```python
# split dataframe in train and test
X = data_2.drop(['Outcome'], axis=1)   # X = independences variables 
y = data_2['Outcome']                  # y = dependence variable

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.3, random_state=192)
```

```python
# 1 Model KNN
model_1 = KNeighborsClassifier(n_neighbors=14, n_jobs=-1)
model_1.fit(X_train, y_train)
y_pred = model_1.predict(X_test)

pd.DataFrame(data=(model_1.score(X_test, y_test),accuracy_score(y_test, y_pred)),
             index=['R', 'Acc'],
             columns=['Mtricas'])
```

```python
# 2 model
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.model_selection import train_test_split


# split dataframe in train and test
X = data_2.drop(['Outcome'], axis=1)   # X = independences variables 
y = data_2['Outcome']                  # y = dependence variable

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.3, random_state=192)


xgb.set_config(verbosity=0)
# create model instance
xgb = XGBClassifier(n_estimators=200, max_depth=100, learning_rate=0.02, objective='binary:logistic')

# fit model
xgb.fit(X_train, y_train)

# make predictions
y_pred = xgb.predict(X_test)
```

```python
from xgboost import plot_tree
fig, ax = plt.subplots(figsize=(20, 15))
plot_tree(xgb, num_trees=0, ax=ax, size="20,15")
plt.savefig('imgs/tree_xgb')
plt.show()
```

```python
pd.DataFrame(data=(xgb.score(X_test, y_test),accuracy_score(y_test, y_pred)),
             index=['R', 'Acc'],
             columns=['Mtricas'])
```

```python
from keras.models import Sequential, model_from_json
from keras.layers import Dense
import matplotlib.pyplot as plt
from keras.models import load_model

# split dataframe in train and test
X = data_2.drop(['Outcome'], axis=1)   # X = independences variables 
y = data_2['Outcome']                  # y = dependence variable

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.3, random_state=192)

tensor = Sequential()
tensor.add(Dense(16, input_dim=4, activation='sigmoid'))
tensor.add(Dense(12, activation='sigmoid'))
tensor.add(Dense(8, activation='sigmoid'))
tensor.add(Dense(12, activation='sigmoid'))
tensor.add(Dense(1, activation='sigmoid'))
tensor.summary()
```

```python
from tensorflow.keras import metrics

# compila a rede e indica o modelo de estimao do erro, optimizao do erro e mtodo/mrito de mensurao do erro.
tensor.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])

# define os arrays de entrada e de sada, o nmero de iteraes de treino, o nmero de entradas por iterao de treino (batch) e mescla as entradas (shuffle)
model_fit_values = tensor.fit(X_train, y_train, epochs=100, batch_size=12, shuffle='true')

# treina o modelo e devolve os scores para print posterior
scores = tensor.evaluate(X_train, y_train)
```

```python
 # mostra o nvel de preciso que o treino atingiu
print("\n%s: %.2f%%" % (tensor.metrics_names[1], scores[1] * 100))

y_pred = tensor.predict(X_test)
y_pred_ = np.zeros(len(y_pred))
for i in range(len(y_pred)):
    if y_pred[i] > 0.5:
        y_pred_[i] = 1
    else:
        y_pred_[i] = 0
```

```python
confusion_matrix(y_test, y_pred_)
```
